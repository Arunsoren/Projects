```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot, plot
import plotly.figure_factory as ff
```

ùêíùêÆùêõùê©ùê•ùê®ùê≠ùê¨

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
#Vertcally stacking
fig, axs = plt.subplots(2)
plt.figure(figsize=(14,6))
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y)
axs[1].plot(x, -y)
```

```python
#Horizontal plot
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(7, 7),
                       gridspec_kw={
                           'width_ratios': [3, 3],
                           'height_ratios': [3, 3],
                       'wspace': 0.4,
                       'hspace': 0.4})
ax[0][0].plot(x, y)
ax[0][1].plot(x, y)
ax[1][0].plot(x, y)
ax[1][1].plot(x, y)
 
plt.show()
```

```python
fig = plt.figure()
fig.set_figheight(6)
fig.set_figwidth(6)
 
ax1 = plt.subplot2grid(shape=(3, 3), loc=(0, 0), colspan=3)
ax2 = plt.subplot2grid(shape=(3, 3), loc=(1, 0), colspan=1)
ax3 = plt.subplot2grid(shape=(3, 3), loc=(1, 2), rowspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1), colspan=1)

 
# plotting subplots
ax1.plot(x, y)
ax1.set_title('ax1')
ax2.plot(x, y)
ax2.set_title('ax2')
ax3.plot(x, y)
ax3.set_title('ax3')
ax4.plot(x, y)
ax4.set_title('ax4')
ax5.plot(x, y)
ax5.set_title('ax5')


plt.tight_layout()
plt.show()
```

ùêÉùê¢ùê¨ùê≠ùê´ùê¢ùêõùêÆùê≠ùê¢ùê®ùêß - We visualize distributions to show the possible values that we can expect to see in a variable, along with how                   likely they are.

sns.distplot - Histograms show the distribution of a single numerical variable.
sns.kdeplot -  KDE plots (or 2D KDE plots) show an estimated, smooth distribution of a single numerical variable (or two                      numerical variables).
sns.jointplot - This command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each                  individual variable.


```python
iris_df = pd.read_csv("iris.csv", index_col="Id")
iris_df.head()
```

```python
iris_df.columns
```

```python
#histogram
plt.figure(figsize=(14,6))
sns.distplot(a = iris_df["Petal Width (cm)"], kde= False)
```

```python
#KDE plot
plt.figure(figsize=(14,6))
sns.kdeplot(data= iris_df["Petal Length (cm)"], shade= True)
```

```python
#pairplot
sns.pairplot(iris_df,hue= "Species" ,  kind="kde")
```

```python
#2D kde plot
plt.figure(figsize=(14, 6))
sns.jointplot(x=iris_df["Petal Length (cm)"], y=iris_df["Sepal Length (cm)"], kind="kde")
```
